```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('insurance.csv')
df.head()
```

```python
x = df.noOfClaims
y = df.totalPayment

plt.xlabel('Number of claims')
plt.ylabel('Total payment')
plt.scatter(x,y)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['noOfClaims']], df.totalPayment)
```

```python
x_vals = 150 * np.random.rand(10,1) - 10
plt.xlabel('Number of claims')
plt.ylabel('Total payment')
plt.scatter(x,y)
plt.plot(x_vals, reg.predict(x_vals))

plt.show()
```
