```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('pizza.csv')
df.head()
```

```python
x = df.annualFranchiseFee
y = df.startUpCost

plt.xlabel('Annual Franchise Fee')
plt.ylabel('Start Up Cost')
plt.scatter(x,y)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['annualFranchiseFee']], df.startUpCost)
```

```python
x_vals = 1000 * np.random.rand(100,1) + 600


plt.xlabel('Annual Franchise Fee')
plt.ylabel('Start Up Cost')
plt.scatter(x,y)
plt.plot(x_vals, reg.predict(x_vals))
plt.show()
```
