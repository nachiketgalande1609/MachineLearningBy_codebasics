Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('sand.csv')
df
```

```python
x = df.medianDiameter
y = df.beachSlopeInDegrees


plt.xlabel('Median Sand Granule Diameter')
plt.ylabel('Beach Slope in degrees')
plt.scatter(x,y)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['medianDiameter']], df.beachSlopeInDegrees)
```

```python
reg.predict([[0.65]])
```

```python
plt.xlabel('Median Sand Granule Diameter')
plt.ylabel('Beach Slope in degrees')
plt.scatter(x,y)
plt.plot(x, reg.predict(df[['medianDiameter']]), color='red')
plt.show()
```
