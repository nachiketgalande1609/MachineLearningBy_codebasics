```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
mtcars = pd.read_csv('mtcars.csv')
mtcars.head()
```

```python
mtcars.info()
```

```python
mtcars.shape
```

Bar Plot

```python
sns.barplot(x='cyl', y='mpg', data=mtcars, palette='rocket')
plt.show()
```

Count Plot

```python
sns.countplot(x='cyl', data=mtcars, color='hotpink')
plt.show()
```

```python
sns.countplot(y='carb', data=mtcars)
plt.show()
```

```python
sns.countplot(x='gear', hue='cyl', data=mtcars)
plt.show()
```

Distribution Plot

```python
sns.displot(mtcars.cyl, bins=10)
plt.show()
```

```python
sns.heatmap(mtcars.corr(), cbar=True, linewidths=0.5)
```

```python
mtcars['cyl'].value_counts()
```

Scatter Plot

```python
sns.scatterplot(x='cyl', y='gear', data=mtcars, hue='carb', palette='rocket')
plt.grid()
```

Pair Plot

```python
sns.pairplot(mtcars, hue='cyl', palette='rocket')
```

Lm Plot

```python
sns.lmplot(x='gear', y='am', data=mtcars)
```

Box Plot

```python
sns.boxplot(x='gear', y='cyl', data=mtcars)
```
