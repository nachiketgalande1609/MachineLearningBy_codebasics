Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
x = np.array([4,9,6,1,2,7,3])
y = np.array([4,6,5,2,4,7,1])

x = x.reshape(-1,1)
y = y.reshape(-1,1)

plt.scatter(x,y)
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(x, y)
```

```python
x_vals = np.linspace(0, 10, 100).reshape(-1,1)
y_vals = reg.predict(x_vals)
```

```python
plt.scatter(x,y)
plt.plot(x_vals, y_vals, color='red')
plt.show()
```
